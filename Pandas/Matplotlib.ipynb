```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

## Line Plot

```python
 plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('LINE PLOT')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth='2')
plt.show()
```

```python
y1 = 2*x
y2 = 3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth='2')
plt.plot(x,y2,color='r',linestyle='-',linewidth='3')

plt.grid()
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*y

plt.subplot(1,2,1)
plt.plot(x,y1,color='yellow')

plt.subplot(1,2,2)
plt.plot(x,y2,color='red')

plt.show()
```

## Bar plot

```python
student = {'bob':78,'raj':90,'srs':89}
```

```python
names = list(student.keys())
```

```python
values = list(student.values())
```

```python
plt.bar(names,values)  # first parameter should be the categorical value and second parameter should be numerical value
plt.show()
```

## horizontal BAR plot

```python
plt.barh(names,values,color='r')
plt.grid(True)
plt.show()
```

## scatter plot

```python
x = [10,20,30,40,50,60]
y = [5,1,7,0,3,2]
```

```python
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.show()
```

```python
plt.scatter(x,y,marker='*',c='g',s=200)
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.show()
```

```python
y2 = [3,6,2,7,4,6]
```

```python
plt.scatter(x,y,marker='*',c='g',s=200)
plt.scatter(x,y2,marker='.',c ='r',s=200)
plt.plot()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*')
plt.plot()

plt.subplot(1,2,2)
plt.scatter(x,y2,marker='.')
plt.plot()
```

## BOX-Plot

```python
one =[1,2,3,4,5,6,7]
two = [1,2,3,4,5,3,1]
three =[8,6,5,6,7,8,9]
data = list([one,two,three])
plt.boxplot(data)
#plt.grid(True)
plt.show()
```

## VIOLIN PLOT

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

## PIE-CHART

```python
cars = ['BMW','AUDI','BENZ','TATA']
quant = [20,29,13,43]

plt.pie(quant,labels=cars)     # first parameter Numerical and then category
plt.show()
```

```python
plt.pie(quant,labels=cars,autopct='%0.1f%%', colors =['yellow','red','white','blue'])
plt.show()
```

## DoughNut-chart

```python
cars = ['BMW','AUDI','BENZ','TATA']
quant = [20,29,13,43]

plt.pie(quant,labels=cars, radius=2) # outer circle
plt.pie([10],colors=['w'],radius=1)  # inner circle w for white radius 1
plt.show()
```
